```python
my_tuple = ("apple", "banana", "cherry")
temp_list = list(my_tuple)
temp_list[1] = "kiwi"
my_tuple = tuple(temp_list)
print(my_tuple)
```

```python
my_tuple = ("apple", "banana", "cherry")
y = list(my_tuple)
y.append("orange")
my_tuple = tuple(y)
print(my_tuple)
```

```python
my_tuple = ("apple", "banana", "cherry")
y = list(my_tuple)
y.remove("apple")
my_tuple = tuple(y)
print(my_tuple)
```
