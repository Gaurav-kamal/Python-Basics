```python
#joining sets
```

```python
x={"a","b","c"}
y={1,2,3,4}
z=x.union(y)
print(z)
```

```python
x={"a","b","c"}
y={1,2,3,4}
z=x | y
print(z)
```

```python
#joining multiple sets
```

```python
x={"a","b","c"}
y={1,2,3,4}
a={"gaurav","rishabh"}
b={"himachal","kashmir"}
c=x.union(y,a,b)
print(c)
```

```python
x={"a","b","c"}
y={1,2,3,4}
a={"gaurav","rishabh"}
b={"himachal","kashmir"}
c=x | y | a | b
print(c)
```
